```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Data/transactions.csv')
```

```python
df
```

```python
df.drop(['transaction_id'], axis=1, inplace=True)
```

```python
df['transaction_time'] = pd.to_datetime(df['transaction_time'])
df['year'] = df['transaction_time'].dt.year
df['month'] = df['transaction_time'].dt.month
df['day'] = df['transaction_time'].dt.day
df['hour'] = df['transaction_time'].dt.hour
```

```python
df.drop(['transaction_time'], axis=1, inplace=True)
```

```python
df
```

```python
X = df[:20000].drop(['is_fraud'], axis=1)
y = pd.DataFrame(df['is_fraud'][:20000].copy())
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns
```

```python
num_features, cat_features
```

```python
num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('encode', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
])

transformer = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features),
], remainder='passthrough')

gb_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('estimator', GradientBoostingClassifier())
])

rfc_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('estimator', RandomForestClassifier())
])
```

```python
gb_pipeline.fit(X_train, y_train)
```

```python
gb_pipeline.score(X_train, y_train), gb_pipeline.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
cross_val_score(gb_pipeline,X_train, y_train, cv=3)
```

```python
preds = cross_val_predict(gb_pipeline, X_train, y_train, cv=3)
```

```python
print(f'Precision score:{precision_score(y_train,preds) * 100}\nRecall score:{recall_score(y_train,preds) * 100}\nF1 score:{f1_score(y_train,preds) * 100}')
```
